# Case study: Multiple linear regression - California housing dataset

The California housing dataset has a real world dataset with 20,640 samples. Each has 8 numerical features.

## Objective
Perform multiple linear regression using all the 8 numerical features.
$y = mx + c$ => $y = B_1 + B_0x$

The prediction of housing is dependent on more than one feature (nearness to market, hospital, police station)
- LinearRegression estimator performs multiple linear regression by default

$y = B_0 + B_1x + B_2y + B_3z$

# Splitting the data for training and testing (train_test_split)

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(california.data, california.target, random_state=11)
```

```
x_train.shape
```

```
x_test.shape
```

## Training the model

* Linear regression tries to use all the features in the dataset.
* It throws an error if any of the features is categorical
* Categorical data must be preprocessed into numerical data or excluded
* NOTE: The scikit-learn bundled datasets are in the correct format

```
from sklearn.linear_model import LinearRegression
```

```
linear_regression = LinearRegression()
```

```
linear_regression.fit(X=x_train, y=y_train)
```

We have separate coefficients for each of the features (stored in coef_)

We have one intercept (stored in intercept_)

* Positive coefficients - median house value increases as the feature value increases
* Negative coefficients - median house value decreases as feature value increases


```
for i, name in enumerate(california.feature_names):
  print(f'{name:<10}: {linear_regression.coef_[i]:.10f}') # name:<10 -- justify to the left and leave 10 characters
```

HouseAge, Population and AveOccup are close to zero. The implication is that they have little or no effect on median house value

```
linear_regression.coef_
```

```
linear_regression.intercept_
```

The equation is therefore:
$MedHouseValue = 0.438MedInc + 0.0092HouseAge -0.107AveRooms + 0.612AveBedrms - 0.0000058Population - 0.0034AveOccup - 0.419Latitude - 0.434Longitude - 36.88$

# Testing the model

```
predicted = linear_regression.predict(x_test)
```

```
expected = y_test
```

```
predicted[:10]
```

```
expected[:10]
```

In classification the predictions are distinct classes that matches the existing classes in the dataset.

In Regression it is difficult to get the exact prediction since we are dealing with continuous outputs

```
df = pd.DataFrame()
```

```
df['Expected'] = pd.Series(expected)
df['Predicted'] = pd.Series(predicted)
```

```
df
```

Plot a scatter plot with the target(expected) on the x axis and predicted on the y axis

```
figure = plt.figure(figsize=(9,9))

axes = sns.scatterplot(data=df, x='Expected', y='Predicted', hue='Predicted', palette='cool', legend=False)

start = min(expected.min(), predicted.min())
end = max(expected.max(), predicted.max())

axes.set_xlim(start, end)
axes.set_ylim(start, end)
line = plt.plot([start, end], [start, end], 'k--') # line of best fit
```

# Regression Model Metrics

Metrics used for Regression estimators:

$R^2$ - coefficient of determination 0.0 (model cannot predict values of the output with accuracy) - 1.0 (perfect prediction)


```
from sklearn import metrics
```

```
metrics.r2_score(expected, predicted)
```

```
linear_regression.score(x_test, y_test)
```

https://scikit-learn.org/stable/modules/linear_model.html 

Using different variants of regression: LinearRegression, ElasticNet, Lasso & Ridge then we pick the best model

```
from sklearn.linear_model import ElasticNet, Lasso, Ridge
```

```
estimators = {
    'LinearRegression': linear_regression,
    'ElasticNet': ElasticNet(),
    'Lasso': Lasso(),
    'Ridge': Ridge()
}
```

We choose the best model by running the estimators using k-fold cross validation (https://machinelearningmastery.com/k-fold-cross-validation/ )

```
from sklearn.model_selection import KFold, cross_val_score
```

```
for estimator_name, estimator_object in estimators.items():
  kfold = KFold(n_splits=10, random_state=11, shuffle=True)
  scores = cross_val_score(estimator=estimator_object, X=california.data, y=california.target, cv=kfold, scoring='r2')
  print(f'{estimator_name:<16}: ' + f'Mean of r2 scores={scores.mean():.4f}')
```

# Conclusion

The best estimator would be LinearRegression and Ridge
